```python
# Pacotes de preparação dos dados
import numpy as np
import pandas as pd
import scipy

# Pacotes gráficos
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

# Pacotes de modelagem
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler

import tensorflow as tf
gpus = tf.config.experimental.list_physical_devices('GPU')
for gpu in gpus:
    tf.config.experimental.set_memory_growth(gpu, True)
from tensorflow import keras
from tensorflow.keras import layers

```

```python
df = pd.read_excel('dados/Todos os resultados da Mega Sena  Rede Loteria.xlsx')
df.columns = df.iloc[0,:]
df = df.drop(0)

colunas = ['Conc', 'Data', '1a_dezena','2a_dezena','3a_dezena','4a_dezena', '5a_dezena','6a_dezena', 'Gan.', 'Premio','Apostas']
df.columns = colunas

df['1a_dezena'] = df['1a_dezena'].astype(int)
df['2a_dezena'] = df['2a_dezena'].astype(int)
df['3a_dezena'] = df['3a_dezena'].astype(int)
df['4a_dezena'] = df['4a_dezena'].astype(int)
df['5a_dezena'] = df['5a_dezena'].astype(int)
df['6a_dezena'] = df['6a_dezena'].astype(int)

df.head()
```

## preparação

```python
df.isnull().sum()
```

```python
df = df.drop(['Premio','Apostas'], axis=1)
```

```python
df['Gan.'] = df['Gan.'].astype(int) 
df['Gan'] = np.where(df['Gan.'] >= 1, 1, 0)
df = df.drop('Gan.', axis=1)
```

## Definição do target e features

```python
# Target (variável resposta)
y_var = 'Gan'
y = df[y_var]

# Features (variáveis explicativas)
x_var_num = ['1a_dezena', '2a_dezena', '3a_dezena', '4a_dezena', '5a_dezena', '6a_dezena']

x = df[x_var_num].to_numpy()

x
```

## Treino e teste

```python
# Divisão em treino e teste
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, 
                                                    test_size=0.30,
                                                    random_state=42)
```

## Definição do NPL

```python
model = keras.Sequential()

# Camada Oculta 1
model.add(layers.Dense(32, 
                       activation='tanh', 
                       kernel_regularizer=keras.regularizers.l1_l2(l1=0.01, l2=0.01),
                       input_dim=x_train.shape[1]))
model.add(layers.Dropout(0.2))

# Camada Oculta 2
model.add(layers.Dense(16, 
                       activation='relu', 
                       kernel_regularizer=keras.regularizers.l1_l2(l1=0.01, l2=0.01),
                       input_dim=x_train.shape[1]))
model.add(layers.Dropout(0.1))


# Camada Oculta 2
model.add(layers.Dense(30, 
                       activation='relu', 
                       kernel_regularizer=keras.regularizers.l1_l2(l1=0.01, l2=0.01),
                       input_dim=x_train.shape[1]))
model.add(layers.Dropout(0.1))


# Camada Saída
model.add(layers.Dense(1, activation='sigmoid'))

# Compilação do Modelo
model.compile(loss='binary_crossentropy', 
              optimizer=keras.optimizers.Adam(),
              metrics=keras.metrics.AUC(name='auc'))

model.summary()
```

### Avaliação da RNA antes do treinamento

```python
score_train, metric_train = model.evaluate(x_train, y_train, verbose=0)
score_test, metric_test = model.evaluate(x_test, y_test, verbose=0)

print('AUC no conjunto de Treino: {:.1%}'.format(metric_train))
print('AUC no conjunto de Teste: {:.1%}'.format(metric_test))
```

### Treinamento da RNA MLP

```python
rna_es = tf.keras.callbacks.EarlyStopping(
    monitor="val_auc",
    min_delta=0.01,
    patience = 20,
    verbose=2,
    mode='max',
    restore_best_weights=True)

historico = model.fit(x=x_train, 
                      y=y_train,
                      epochs=100,
                      batch_size=32,
                      validation_split=0.1,
                      verbose=1,
                      callbacks=[rna_es])
```

### Avaliação do RNA

```python
score_train, metric_train = model.evaluate(x_train, y_train, verbose=0)
score_test, metric_test = model.evaluate(x_test, y_test, verbose=0)

print('AUC no conjunto de Treino: {:.1%}'.format(metric_train))
print('AUC no conjunto de Teste: {:.1%}'.format(metric_test))
```

```python
auc_treino = list(historico.history.keys())[1]
auc_valid = list(historico.history.keys())[3]

# Gráfico com a evolução do treinamento
fig, ax = plt.subplots(figsize=(8, 8))
plt.style.use('ggplot')
plt.plot(historico.history[auc_treino])
plt.plot(historico.history[auc_valid])
plt.title('Desempenho do Modelo')
plt.ylabel('AUC')
plt.xlabel('Épocas')
plt.legend(['Treino', 'Validação'], loc='upper left')
plt.show()
```

### melhorando os hiperparamentros

```python
def treina_rna(x_train, y_train, x_test, y_test, nnc1, nnc2, nnc3):

    # Definição do Modelo RNA
    model = keras.Sequential()

    # Camada Oculta 1
    model.add(layers.Dense(nnc1, 
                           activation='relu', 
                           kernel_regularizer=keras.regularizers.l1_l2(l1=0.01, l2=0.01),
                           input_dim=x_train.shape[1]))
   

    # Camada Oculta 2
    model.add(layers.Dense(nnc2, 
                           activation='relu', 
                           kernel_regularizer=keras.regularizers.l1_l2(l1=0.01, l2=0.01)))
    
    # Camada Oculta 3
    model.add(layers.Dense(nnc3, 
                           activation='relu', 
                           kernel_regularizer=keras.regularizers.l1_l2(l1=0.01, l2=0.01)))
    


    # Camada Saída
    model.add(layers.Dense(1, activation='sigmoid'))

    # Compilação do Modelo
    model.compile(loss='binary_crossentropy', 
                  optimizer=keras.optimizers.Adam(),
                  metrics=keras.metrics.AUC(name='auc'))


    # Treinamento do Modelo RNA
    rna_es = tf.keras.callbacks.EarlyStopping(
    monitor="val_auc",
    min_delta=0.01,
    patience = 5,
    verbose=0,
    mode='max',
    restore_best_weights=True)

    historico = model.fit(x=x_train, 
                          y=y_train,
                          epochs=100,
                          batch_size=32,
                          validation_split=0.1,
                          verbose=0,
                          callbacks=[rna_es])


    
    # Avaliação do Desempenho do Modelo RNA
    score_train, metric_train = model.evaluate(x_train, y_train, verbose=0)
    score_test, metric_test = model.evaluate(x_test, y_test, verbose=0)

    
    return model, (metric_train, metric_test)
```

```python
%%time
lst_models = list()
lst_nnc1 = list()

lst_nnc2 = list()
lst_nnc3 = list()
auc_train = list()
auc_test = list()

id_model = 0
for nnc1 in range(24, 32):
    #for dc1 in np.arange(0, 0.2, 0.1):
        for nnc2 in range(24, 32):
            for nnc3 in range(30, 32):
                print(f'Treinando Modelo {id_model}')
                lst_nnc1.append(nnc1)
                #lst_dc1.append(dc1)
                lst_nnc2.append(nnc2)
                lst_nnc3.append(nnc3)
                modelo, auc = treina_rna(x_train, y_train, 
                                         x_test, y_test,
                                         nnc1, nnc2,nnc3)
                lst_models.append(modelo)
                auc_train.append(auc[0])
                auc_test.append(auc[1])
                id_model += 1
```

```python
busca_rna = pd.DataFrame(list(zip(lst_nnc1, lst_nnc2,lst_nnc3, auc_train, auc_test, lst_models)), 
                        columns=['Num Neuronios C1',
                                 'Num Neuronios C2',
                                 'NumNeuronios C3',
                                 'AUC Train', 'AUC Test', 
                                 'Modelo'
                                ])
busca_rna['Diferença'] = busca_rna['AUC Test'] - busca_rna['AUC Train']
busca_rna
```

```python
def analise_hiperparametro(hp, busca):
    matplotlib.use('module://ipykernel.pylab.backend_inline')
    sns.set()

    fig, axes = plt.subplots(1, 2, figsize=(9, 4))
    fig.suptitle(t='Análise do Hiperparâmetro: '+hp, fontsize=12)

    sns.lineplot(ax=axes[0], x=busca[hp], y=busca['AUC Train'])
    sns.lineplot(ax=axes[0], x=busca[hp], y=busca['AUC Test'])
    axes[0].legend(labels=['Treino', 'Teste'])

    sns.lineplot(ax=axes[1], x=busca[hp], y=busca['Diferença'])

    fig.tight_layout()    
```

```python
import numpy as np

def prever_numeros_megasena(modelo, quantidade_numeros=6, limiar_probabilidade=0.5):
    probabilidade_ganhar = 0
    
    while probabilidade_ganhar <= limiar_probabilidade:
        # Gerar probabilidades fictícias (substitua isso pelo output real do seu modelo)
        probabilidades_ficticias = np.random.rand(60)

        # Normalizar as probabilidades para somarem 1
        probabilidades_normalizadas = probabilidades_ficticias / probabilidades_ficticias.sum()

        # Escolher números com base nas probabilidades normalizadas
        numeros_escolhidos = np.random.choice(range(1, 61), size=quantidade_numeros, replace=False, p=probabilidades_normalizadas)

        # Fazendo a predição - Probabilidades
        entrada = np.array([numeros_escolhidos])
        probabilidades = modelo.predict(entrada)

        # Obtendo a probabilidade da classe 1 (chance de ganhar)
        probabilidade_ganhar = probabilidades[0][0]    

    return numeros_escolhidos, probabilidade_ganhar

# Exemplo de uso com um loop para gerar 5 sequências de números ganhadores com probabilidade superior a 50%
quantidade_sequencias = 5
for _ in range(quantidade_sequencias):
    numeros_ganhadores, probabilidade_ganhar = prever_numeros_megasena(modelo_carregado)
    print("Números previstos da Mega Sena:", numeros_ganhadores, "-> Probabilidade:", round(probabilidade_ganhar * 100, 2), "%")
```
